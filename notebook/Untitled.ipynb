```python
import json
```

```python
format_1_answer_1 = []
format_1_answer_0 = []
format_0_answer_0 = []
with open('../data/hendrycks_math/evaluation_results.jsonl', 'r') as f:
    for line in f:
        data = json.loads(line)
        rewards = data['rewards']
        if rewards['format_reward'] == 1 and rewards['answer_reward'] == 1:
            format_1_answer_1.append(data)
        elif rewards['format_reward'] == 1 and rewards['answer_reward'] == 0:
            format_1_answer_0.append(data)
        elif rewards['format_reward'] == 0 and rewards['answer_reward'] == 0:
            format_0_answer_0.append(data)
```

```python
with open('../data/hendrycks_math/evaluation_results_1_1.jsonl', 'a') as f:
    for data in format_1_answer_1:
        f.write(json.dumps(data, ensure_ascii=False) + "\n")

with open('../data/hendrycks_math/evaluation_results_1_0.jsonl', 'a') as f:
    for data in format_1_answer_0:
        f.write(json.dumps(data, ensure_ascii=False) + "\n")

with open('../data/hendrycks_math/evaluation_results_0_0.jsonl', 'a') as f:
    for data in format_0_answer_0:
        f.write(json.dumps(data, ensure_ascii=False) + "\n")
```
