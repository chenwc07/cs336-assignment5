```python
from cs336_alignment.utils.utils import tokenize_prompt_and_output
from cs336_alignment.utils.sft_dataloader import SFTTrainDataset
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("/data1/chenweicong/models/Qwen/Qwen2.5-Math-1.5B")
```

```python
train_dataset = SFTTrainDataset(
    data_path='data/hendrycks_math/sft.jsonl',
    prompt_template_path='cs336_alignment/prompts/r1_zero.prompt',
)
```

```python
length = []
for i in range(1000):
    batch_prompts, batch_answers = train_dataset.get_batch(2)
    tokenized_batch = tokenize_prompt_and_output(
        batch_prompts,
        batch_answers,
        tokenizer,
    )
    input_ids = tokenized_batch["input_ids"]
    length.append(input_ids.shape[1])
```

```python
max(length)
```

```python
from cs336_alignment.drgrpo_grader import r1_zero_reward_fn
import json
```

```python
good_answer = []
bad_answer = []
with open('data/hendrycks_math/sft_short.jsonl', 'r') as f:
    for line in f:
        data = json.loads(line)
        reward = r1_zero_reward_fn(data['ds_answer'], data['solution'])
        if reward['reward'] == 1:
            good_answer.append(data)
        else:
            bad_answer.append(data)
```

```python
len(good_answer)
```

```python
len(bad_answer)
```

```python
with open('data/hendrycks_math/sft_short_filter.jsonl', 'w') as f:
    for data in good_answer:
        f.write(json.dumps(data) + '\n')
```
